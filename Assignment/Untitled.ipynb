```python
#Library import

import numpy as np

#OpenCV for face detection
import cv2

#os to work with images
import os

#The glob module finds all the pathnames matching a specified pattern
import glob

import dlib
```

```python
imgs_dir = "C:/Users/ching/Desktop/Assignment/dataset"
imgs_path = os.path.join(imgs_dir,'*png')
imgs = glob.glob(imgs_path)
imgs
```

```python
face_cascade = cv2.CascadeClassifier('C:/Users/ching/Anaconda3/Library/etc/haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('C:/Users/ching/Anaconda3/Library/etc/haarcascades/haarcascade_eye.xml')
```

```python
for i in range (0, 5000):
    img = cv2.imread(imgs[i])
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    if (type(faces) is np.ndarray):
        os.remove(imgs[i])

```

```python
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)

cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
